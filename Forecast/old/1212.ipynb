```python
import matplotlib.pyplot as plt
import numpy as np

# Данные
data = [24, 0, 0, 0, 0, 0, 0, 18, 50	,99,	87,	47
]
x = np.arange(1, len(data)+1)

# Линейная регрессия (прямая)
a, b = np.polyfit(x, data, 1)  # коэффициенты линии y = a*x + b
y_fit = a * x + b

# Построение графика
plt.figure(figsize=(10,5))
plt.plot(x, data, 'o', label='Исходные точки')
plt.plot(x, y_fit, '-', color='red', label='Линейная регрессия')
plt.xlabel('Номер точки')
plt.ylabel('Значение')
plt.title('Линейная аппроксимация')
plt.grid(True)
plt.legend()
plt.show()
```
